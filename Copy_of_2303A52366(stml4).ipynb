1.read the data with pandas and find features and target variables

```python
import pandas as pd
```

```python
d = pd.read_csv('/content/train.csv')
print(d)
```

```python
y = d['price_range']
print(y)
```

```python
x=d.drop('price_range',axis=1)
print(x)
```

```python
df=x-x.min()/x.max()-x.min()
print(df)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df,y,test_size=0.2,random_state=60)
print(xtrain)
print(xtest)
print(ytrain)
print(ytest)
```
