Question 1:
From the above data:

Read the data with pandas and find features and target variables
Plot a graph between features and target
Find Best fit line using linear regression.
FInd MSE, MAE, for test size (20,25)

```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, mean_absolute_error

df = pd.read_csv('/content/Salary_Data.csv')

X= df [['YearsExperience']]

y= df ['Salary']

plt.scatter (X, y, color= 'blue', label ='Experience vs Salary')

plt.title('Experience vs Salary')

plt.xlabel('Years of Experience')

plt.ylabel('Salary')

plt.legend()

plt.grid(True)

plt.show()

for test_size in [0.20, 0.25]:
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size, random_state=42)
  model = LinearRegression()
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  mse = mean_squared_error(y_test, y_pred)
  mae = mean_absolute_error(y_test, y_pred)
  print(f"Test Size: {int(test_size+100)}%")
  print (f"MSE: {mae}")
  print (f"MAE: {mae}\n")
  plt.scatter(X_test, y_test, color='blue', label ='Test Data')
  plt.plot(X_test, y_pred, color='red', label='Best Fit Line')
  plt.title(f'Linear Regression with Test Size {int(test_size*100)}%')
  plt.xlabel('Years of Experience')
  plt.ylabel('Salary')
  plt.legend()
  plt.grid(True)
  plt.show()
```
